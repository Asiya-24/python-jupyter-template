### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.

import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

plt.figure(figsize=(8, 6))  # Optional: Adjust figure size

plt.hist(data, bins=range(1, 7), edgecolor='black', color='skyblue')  # Customize bins, color, and edge

plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.xticks(range(1, 6))  # Set x-axis ticks to match the data values

plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add gridlines

plt.show()
```
